```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def con_tup(list):
    return tuple(list)
```

```python
d_column = df.iloc[:,5] #Do you think that transgenders should study?
```

```python
d_column.value_counts()
```

```python
def siz(column):
    l = []
    for i in range(len(column.value_counts())):
        l.append(column.value_counts()[i])
    return l
```

```python
sizes = siz(d_column)
```

```python
labels = con_tup(d_column.value_counts().index.tolist())
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
